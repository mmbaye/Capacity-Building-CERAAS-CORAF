# Importing Python Libraries
In Python, libraries (also known as modules) are collections of pre-written code that provide additional functionality to your programs. To use a library in your Python code, you need to import it. Here are a few ways to import libraries in Python:
``import library_name``

1. Example: **import cv2** This imports the entire library, and you can access its functions and classes using the library name as a prefix. For example, to use the **_imshow()_** function from the cv2 library, you would write **_cv2.imshow()_**.

Import the entire library with an alias, such as  ``import library_name as alias
``
2. Example: ``import numpy as np`` This imports the library with a shorter alias, making it easier to use. You can then use the alias to access functions and classes from the library. For example, np.array([1, 2, 3]) instead of numpy.array([1, 2, 3]).

Import specific functions/classes from a library: ``from library_name import function_name, class_name``
3. Example: ``from matplotlib import pyplot as plt`` This imports only the specified function(s) or class(es) from the library. You can directly use the imported names without the library prefix.
